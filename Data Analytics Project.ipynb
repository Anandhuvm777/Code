# Exporting dataframe to SQL - using SQLALCHEMY library

```python
import sqlalchemy as sal
```

```python
# creating an engine to connect with SQL
# instead of mssql if we are connecting with mysql or postgresql, write them.
# MSI\\SQLEXPRESS/HRDB?  -> server name
# HRDB-> database name

engine = sal.create_engine('mssql://MSI\\SQLEXPRESS/HRDB?driver=ODBC+Driver+18+for+SQL+Server&TrustServerCertificate=yes')
```

The create_engine function creates an SQLAlchemy Engine object, which establishes a connection to the SQL Server database. The connection string specifies the database type (mssql), the server name (MSI\SQLEXPRESS), the database name (HRDB), and the driver (ODBC DRIVER 18 FOR SQL SERVER).

```python
conn=engine.connect()
conn
```

```python
df1.to_sql('orders_data',con=conn,index=False,if_exists='replace') # or if_exists = 'append'
```

This line uses the Pandas to_sql method to write the DataFrame df to the table named orders_data in the SQL Server database. The con parameter specifies the connection object, index=False indicates that the DataFrame's index should not be written as a separate column in the database table, and if_exists='replace' indicates that if the table already exists, it should be replaced. Alternatively, you can use if_exists='append' to add new rows to the existing table.
